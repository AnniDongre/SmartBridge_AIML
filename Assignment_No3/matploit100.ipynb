```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set style
sns.set_style("darkgrid")

# Generate sample data
np.random.seed(42)
x = np.linspace(1, 10, 50)
y = np.sin(x) + np.random.normal(scale=0.2, size=len(x))
categories = np.random.choice(['A', 'B', 'C', 'D'], size=50)

# Increase figure size
plt.figure(figsize=(20, 10))

# 1. Line Plot
plt.subplot(2, 5, 1)
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title("Line Plot", fontsize=14)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

# 2. Scatter Plot
plt.subplot(2, 5, 2)
plt.scatter(x, y, c=np.random.rand(50), cmap='viridis', edgecolors='k')
plt.colorbar(label="Random Color Scale")
plt.title("Scatter Plot", fontsize=14)

# 3. Bar Chart
category_counts = pd.Series(categories).value_counts()
plt.subplot(2, 5, 3)
category_counts.plot(kind='bar', color=['red', 'blue', 'green', 'purple'])
plt.title("Bar Chart", fontsize=14)

# 4. Histogram
data = np.random.randn(1000)
plt.subplot(2, 5, 4)
plt.hist(data, bins=30, color='orange', alpha=0.7, edgecolor='black')
plt.title("Histogram", fontsize=14)

# 5. Box Plot
plt.subplot(2, 5, 5)
sns.boxplot(data=data, orient="h", color="cyan")
plt.title("Box Plot", fontsize=14)

# 6. Pie Chart
plt.subplot(2, 5, 6)
sizes = [25, 35, 20, 20]
labels = ['Apple', 'Banana', 'Cherry', 'Date']
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['r', 'g', 'b', 'y'])
plt.title("Pie Chart", fontsize=14)

# 7. Heatmap
plt.subplot(2, 5, 7)
matrix_data = np.random.rand(5, 5)
sns.heatmap(matrix_data, annot=True, cmap='coolwarm')
plt.title("Heatmap", fontsize=14)

# 8. Violin Plot
plt.subplot(2, 5, 8)
sns.violinplot(data=data, color="lightblue")
plt.title("Violin Plot", fontsize=14)

# 9. Area Plot
plt.subplot(2, 5, 9)
plt.fill_between(x, y, color="skyblue", alpha=0.5)
plt.title("Area Plot", fontsize=14)

# 10. Step Plot
plt.subplot(2, 5, 10)
plt.step(x, y, color="magenta", where="mid")
plt.title("Step Plot", fontsize=14)

plt.tight_layout()
plt.show()
```
