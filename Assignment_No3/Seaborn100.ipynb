```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample DataFrame
np.random.seed(42)
data = {
    "A": np.random.randint(1, 100, 50),
    "B": np.random.rand(50) * 100,
    "C": np.random.randint(1, 500, 50),
    "D": np.random.choice(["X", "Y", "Z"], 50),
    "E": np.random.choice(["Male", "Female"], 50)
}
df = pd.DataFrame(data)

# Create a 3x3 subplot grid for the first 9 plots
fig, axes = plt.subplots(3, 3, figsize=(15, 12))

# 1. Histogram
sns.histplot(df["A"], bins=10, kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Histogram of A")

# 2. Box Plot
sns.boxplot(x=df["D"], y=df["B"], ax=axes[0, 1])
axes[0, 1].set_title("Box Plot of B by Category D")

# 3. Scatter Plot
sns.scatterplot(x=df["A"], y=df["B"], hue=df["E"], ax=axes[0, 2])
axes[0, 2].set_title("Scatter Plot of A vs. B")

# 4. Bar Plot
sns.barplot(x=df["D"], y=df["C"], ax=axes[1, 0])
axes[1, 0].set_title("Bar Plot of C by Category D")

# 5. Line Plot
sns.lineplot(x=range(len(df)), y=df["A"], ax=axes[1, 1])
axes[1, 1].set_title("Line Plot of A over Index")

# 6. Violin Plot
sns.violinplot(x=df["E"], y=df["C"], ax=axes[1, 2])
axes[1, 2].set_title("Violin Plot of C by Gender E")

# 7. Heatmap (Only Numeric Columns)
sns.heatmap(df.select_dtypes(include="number").corr(), annot=True, cmap="coolwarm", ax=axes[2, 0])
axes[2, 0].set_title("Heatmap of Numeric Features")

# 8. KDE Plot
sns.kdeplot(x=df["B"], y=df["C"], fill=True, cmap="Blues", ax=axes[2, 1])
axes[2, 1].set_title("KDE Plot of B vs. C")

# 9. Count Plot
sns.countplot(x=df["D"], hue=df["E"], ax=axes[2, 2])
axes[2, 2].set_title("Count Plot of D by Gender E")

# Adjust layout
plt.tight_layout()

# 10. Pair Plot (Separate Figure)
sns.pairplot(df, hue="E")
plt.suptitle("Pair Plot of Dataset", y=1.02)

plt.show()
```
